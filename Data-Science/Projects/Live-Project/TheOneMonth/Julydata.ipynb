## This code gives the total amount spent on various Sub-Cat

```python
loop_output ={}
for i in unique_values:
    key = f'{i}'
    details=salesData[salesData["Expense Details"].str.startswith(i)]["Amount"].array.sum()
    loop_output[key]=details
```

```python
cashInData["Details"] = np.where(cashInData["Details"] == "Counter cash ",
                     "Counter cash",
                     cashInData["Details"])
```

```python
loop_output_incash ={}
for i in cashInData["Details"].unique():
    key = f'{i}'
    details= cashInData[cashInData["Details"].str.startswith(i)]["Amount"].array.sum()
    loop_output_incash[key]=details
```
